```python
import pandas as pd
from matplotlib import pyplot as pt
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
put = df.drop("Survived",axis="columns")
put
```

```python
target = df["Survived"]
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()
```

```python
put["sex_n"] = le_Sex.fit_transform(put["Sex"])
```

```python
put
```

```python
put_n = put.drop(["Name","Sex","Ticket","Cabin","Embarked","SibSp","Parch"],axis="columns")
```

```python
put_n
```

```python
put_n["Age"].value_counts()
```

```python
put_n.fillna(value=put_n["Age"].mean(), inplace=True)
```

```python
put_n
```

```python
put_c = put_n
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(put_c,target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict([[1,3,22.0,7.2500,1]])
```
